### 4.2.3. Cost Function


 - **style transfer의 두 가지  비용 함수**
 ![](./Images/week4-4-2_image/9.png)   
 <br/>  

   - Content cost Function: 내용 이미지 C(Content)와 생성된 이미지 G(Gnerated)와의 비슷한 정도를 측정
   - Style cost function:  스타일 이미지 S(Style)와 생성된 이미지 G(Gnerated)와의 비슷한 정도를 측정

<br/>     

- **Cost Function 정의**  
  - $J{(G)} = \alpha J_{content}{(C,G)} + \beta J_{style}{(S,G)}$
  - 상대적 가중치를 두기 위한 두 하이퍼파라미터 α, β  

<br/>  

- **생성되는 이미지 G(generated) 찾기**   

![](./Images/week4-4-2_image/10.png)   

 1. 생성 이미지 G를 무작위로 초기화 (위에서 흰 노이즈의 첫 이미지)
 2. Cost Function인 J(G)를 정의 
 3. Gradient descent로 J(G)를 최소화   


 →  픽셀값이 업데이트되어 내용 이미지와 스타일 이미지가 융합된 이미지가 생성
 




### 4.2.4. Content Cost Function
- **content cost function을 정의하는 법**
  - content cost를 계산하기 위해서 은닉층 중 하나를 선택
    - 보통 은닉층 중간 정도에 있는 레이어를 선택함
  - 미리 학습된 네트워크를 활용 ex) vgg
  - $a^{[1](C)}$와 $a^{[1](G)}$를 각각 원본과 생성된 이미지 대한 activation라고 할 때, 두 값이 유사하다면 두 이미지가 유사한 내용을 가지고 있다고 할 수 있음
  - $J_{content}(C, G) = \alpha * ||a^{[1](C)} - a^{[1](G)}||^{2}$로 정의

### 4.2.5.  Style Cost Function
- **Image Style**  
<br/>
![](./Images/week4-4-2_image/11.jpg)  
<br/>
  - Style의 정의: 한 레이어 안에서 '다른 채널 간 활성 정도의 상관 관계'
  - 한 레이어의 활성 블록을 $n_h$ \* $n_w$ \* $n_c$라 할때, 채널 i와 채널 j의 모든 ( $n_h$ , $n_w$) 위치에서 Style을 표현 

<br/>
<br/>

- **Intuition about style of an image**  
<br/>
![](./Images/week4-4-2_image/12.jpg)      
<br/>
  - CNN에서 각 layer의 output(activation)은 input을 여러 channel로 convolution한 image들의 집합
  - 빨강 테두리의 activation이 빨강 Channel과 관계되고 '세로 선'들을 특징적으로 인식하는 것,  
    노랑 테두리의 activation은 노랑 Channel과 관계되고 '붉은 계열의 경계선'을 특징적으로 인식하는 것이라 할 때,    
    이 두가지 activation의 상관관계로 Style을 정의  
    → 세로 선이 존재 할 때 붉은 계열도 존재하는지 
  - 모든 channel에 의한 activation들의  correlation로 Style Matrix를 구성  

<br/>
<br/>

- **Style Matix**
<br/>
![](./Images/week4-4-2_image/13.jpg)  
<br/>
  - G(Gram matrix 또는 Style matix)는 correlation matrix
  - 레이어 l, 높이 i, 너비 j, 채널 k(1,...,nc) 일 때 nc x nc 차원의 제곱 매트릭스   
  - G는 채널 k에서의 활성화와 채널 k'에서의 활성화가 얼마나 상관 관계가 있는지를 측정한 것으로,   
     (k,k')값은 k번째와 k'번째의 channel들로 conv한 것에서 같은 위치($n_h$, $n_w$)의 activation값들을 곱해서 합한 것  
     (엄밀하게는 평균을 빼지 않고 단지 이 원소들을 곱하기 때문에 비정규화 교차 공분산)  

<br/>
<br/>
          
- **Style cost function**
<br/>
![](./Images/week4-4-2_image/14.jpg)  
<br/>
  - 목표는 S(Style)와 G(Generated)의 "Gram matrix"차이를 줄이는 것
      - Style loss function =  Gram 행렬 간 차이의 제곱   
        (프로베니우스 노름(Frobenius Norm) 이용, 정규화 상수로 나눈 후 하이퍼파라미터 $beta$를 곱할 것)
      - 전체 Style cost function = 모든 Layer에서의 λ^[l]\*(Style cost)의 합 
      - 전체 Cost function =  α\*(C, G의 전체 cost) + β\*(S, G의 전체 style cost)
  
  - Style Cost Function은 Content Cost Function과 다른 방식이지만 마찬가지로  pre-trained된 CNN 모델을 사용(ex) vgg

### 4.2.6. 1D and 3D Generalizations
